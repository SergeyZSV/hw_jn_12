```python
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
root = os.getcwd()
filename = os.path.join(root, 'company_sales_data.csv')
filename
```

```python
data = pd.read_csv(filename)
data.head()
```

```python
months = data['month_number'].tolist()
```

```python
f_cream = data['facecream'].tolist()
f_wash = data['facewash'].tolist()
t_paste = data['toothpaste'].tolist()
b_soap = data['bathingsoap'].tolist()
shamp = data['shampoo'].tolist()
moist = data['moisturizer'].tolist()
```

```python
total_units = data['total_units'].tolist()
total_profit = data['total_profit'].tolist()
```

### 1. Прочитайте все данные о продажах продуктов и покажите их с помощью многострочного графика

```python
plt.plot(months, f_cream,
        label = 'Face cream')

plt.plot(months, f_wash,
        label = 'Face wash')

plt.plot(months, t_paste,
        label = 'Tooth paste')

plt.plot(months, b_soap,
        label = 'Bathing soap')

plt.plot(months, shamp,
        label = 'Shampoo')

plt.plot(months, moist,
        label = 'Moisturizer')


plt.title('Sales data')
plt.legend(loc = 'upper left')
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales')


plt.show()
```

### 2. Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.

```python
b_soap
```

```python
plt.bar(months, b_soap,
       label = 'Bathing soap sales',
       color = 'g', width = 0.5)

plt.title('Bathing soap sales')
plt.legend(loc = 'upper left')
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales')


plt.show()
```

### 3. Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

```python
f_cream_total = sum(f_cream)
f_wash_total = sum(f_wash)
t_paste_total = sum(t_paste)
b_soap_total = sum(b_soap)
shamp_total = sum(shamp)
moist_total = sum(moist)
sales = [f_cream_total, f_wash_total, t_paste_total, b_soap_total, shamp_total, moist_total]
```

```python
labels = 'Face cream', 'Face wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer'
plt.pie(sales,
       labels = labels,
       autopct = '%1.1f%%')
plt.title('Annual sales')

plt.show()
```
